# Epidemiology Dataframes

### psycopg2  - Library to connect PostgreSQL with Python
### pandas - Python Data Analysis Library

```python
import psycopg2 as pg
import pandas.io.sql as psql
import pandas as pd
```

#### Pandas configuration

```python
pd.set_option("display.max_rows", 500)
```

#### Create the connection variables (RMCAB & Epidemiology)
#### Create the dataframes (RMCAB Kennedy & Puente Aranda) and Epidemiology DB

```python
connection_rmcab = pg.connect("dbname=rmcab_db_final user=postgres")
connection_epidemiology = pg.connect("dbname=epidemiology_db user=postgres")

rmcab_ptearanda_dataframe = psql.read_sql("SELECT * FROM \"RMCAB_PTE_ARANDA\"", connection_rmcab)
temp_ptearanda_dataframe = psql.read_sql("SELECT * FROM \"TEMP_PTE_ARANDA\"", connection_rmcab)
rmcab_kennedy_dataframe = psql.read_sql("SELECT * FROM \"RMCAB_KENNEDY\"", connection_rmcab)
temp_kennedy_dataframe = psql.read_sql("SELECT * FROM \"TEMP_KENNEDY\"", connection_rmcab)
rmcab_carvajal_dataframe = psql.read_sql("SELECT * FROM \"RMCAB_CARVAJAL\"", connection_rmcab)
temp_carvajal_dataframe = psql.read_sql("SELECT * FROM \"TEMP_CARVAJAL\"", connection_rmcab)

epidemiology_ira_daily_dataframe = psql.read_sql("SELECT * FROM \"IRA_DAILY_VIEW\"", connection_epidemiology)
```

#### Merge the dataframes RMCAB (Puente Aranda & Kenneddy) and Epidemiology_db (IRA_DAILY_VIEW)

```python
pte_aranda_full = pd.merge(rmcab_ptearanda_dataframe,
                           temp_ptearanda_dataframe,
                           how='outer',
                           on='FECHA')

pte_aranda_full = pd.merge(pte_aranda_full,
                           epidemiology_ira_daily_dataframe,
                           how='outer',
                           on='FECHA')

kennedy_full = pd.merge(rmcab_kennedy_dataframe,
                          temp_kennedy_dataframe,
                          how='outer',
                          on='FECHA')

kennedy_full = pd.merge(kennedy_full,
                          epidemiology_ira_daily_dataframe,
                          how='outer',
                          on='FECHA')

carvajal_full = pd.merge(rmcab_carvajal_dataframe,
                          temp_carvajal_dataframe,
                          how='outer',
                          on='FECHA')

carvajal_full = pd.merge(carvajal_full,
                          epidemiology_ira_daily_dataframe,
                          how='outer',
                          on='FECHA')
```
